```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
from langchain_core.prompts import ChatPromptTemplate

llm = ChatGoogleGenerativeAI(model = 'gemini-2.5-flash-preview-09-2025', temperature=0.3, max_tokens = 200)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ('system', 'You are an senior AI Architect'),
    ('human', 'Explain this {topic} in simple terms with examples')
])
chain  = prompt | llm
chain.invoke({'topic':'AWS Bedrock'})
```
